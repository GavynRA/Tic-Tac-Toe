```python
def marker_choice():
    '''Function to determine what symbol player 1 wants to be.'''
    
    #initalizing the symbol and what values are acceptable
    marker1 = ' '
    acceptable_markers = ['X', 'O']
    
    #loop until an acceptable input has been given
    while marker1 not in acceptable_markers:
        
        marker1 = input('Player 1 please pick your marker "X" or "O" ')
        
        #simple error message
        if marker1 not in acceptable_markers:
            print('Please choose either "X" or "O"')
    
    return marker1
```

```python
def board_print(game_list):
    '''Function to print the board of the curent gamestate.'''

    #clear the screen
    clear_output()
    
    #board is 17 characters across
    print(f'  {game_list[7]}  |  {game_list[8]}  |  {game_list[9]}  ')
    print(f'-----------------')
    print(f'  {game_list[4]}  |  {game_list[5]}  |  {game_list[6]}  ')
    print(f'-----------------')
    print(f'  {game_list[1]}  |  {game_list[2]}  |  {game_list[3]}  ')
    
```

```python
def player_input(game_list, empty_positions):
    '''
    Function to ask the user where they would like to place their marker. 
    An empty position list is being passed in so only positions that are empty can be used.
    Note need to remove the value after filling the game list!!!
    '''
    #initializing
    input_num = ' '
    
    #loop unitl an acceptable input is given (is a number between 1 and 9 that has not been filled in)
    while input_num.isdigit() == False or int(input_num) not in range(1,10) or int(input_num) not in empty_positions:
    
        input_num = input('Where would you like to go? (1-9)')
        
        #error messages for correponding incorrect inputs
        if input_num.isdigit() == False:
            print('Please input a number')
            
        elif int(input_num) not in range(1,10):
            print('The number must be between 1 and 9')
            
        elif int(input_num) not in empty_positions:
            print(f'Position {int(input_num)} is already filled, nice try')
            
    return int(input_num)
```

```python
def game_condition(game_list):
    '''
    Function to check if the game has been won or not.
    If the game has been won the corresponding marker will be returned with an empty string being returned if no-one has won.
    '''
    win_values = [(1,2,3),(4,5,6),(7,8,9),(1,4,7),(2,5,8),(3,6,9),(1,5,9),(3,5,7)]
    
    for a,b,c in win_values:
            
        if game_list[a] == game_list[b] == game_list[c] == 'X':
            return 'X'
        elif game_list[a] == game_list[b] == game_list[c] == 'O':
            return 'O'
        else:
            pass
    return 'cont'    
```

```python
def continue_check():
    #initialize
    answer = ' '
    #loop until an acceptable input has been given
    while answer not in ['Y','N']:
        
        answer = input('Would you like to play again? (Y or N)')
        
        #simple error message
        if answer not in ['Y','N']:
            print("Please choose either 'Y' or 'N'.")
            
    #return boolean for game_cont        
    if answer == 'Y':
        return True
    else:
        return False
```

```python
#import the clear screen function for notebooks
from IPython.display import clear_output
#import the random library
import random

#initialize values
game_cont = True
player_markers = {'player 1': ' ', 'player 2': ' '}
first_player = random.choice(['player 1', 'player 2'])

#choose the markers for the players
player_markers['player 1'] = marker_choice()
if player_markers['player 1'] == 'X':
    player_markers['player 2'] = 'O'
else:
    player_markers['player 2'] = 'X'
    


while game_cont == True:
    
    #initialize values for the start of a fresh game
    game_list = ['#', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
    empty_positions = list(range(1,10))
    gamestate = 'cont'
    active_player = first_player


    #ask players to place markers while the game has not been won or drawn
    while gamestate == 'cont' and len(empty_positions) > 0:

        #ask for an input from the active player
        print(f'{active_player} where would you like to go?')
        placement = player_input(game_list,empty_positions)

        #change the game list
        game_list[placement] = player_markers[active_player]

        #print the new board
        board_print(game_list)

        #check if the game has been won by reassigning the gamestate value
        gamestate = game_condition(game_list)
        
        #remove the selected position from the available list.
        #This had to be done after the gamestate check or the loop would immediately break on len(empty_positions) = 0
        empty_positions.remove(placement)

        #switch the players around
        if active_player == 'player 1':
            active_player = 'player 2'
        else:

            active_player = 'player 1'



    #announce who has won
    if gamestate == 'cont':
        print('Game over, a draw has been reached!')
    elif gamestate == player_markers['player 1']:
        print("Congratulations player 1, you've won")
    elif gamestate == player_markers['player 2']:
        print("Congratulations player 2, you've won")
    else:
        print('error 4 : no-one won but no draw?')

    # ask if want to play again
    game_cont = continue_check()
    
    #change the first player to make it fair
    if first_player == 'player 1':
        first_player = 'player 2'
    else:
        first_player = 'player 1'

    
print('Thanks for playing')
```
